```python
class SelectItem(torch.nn.Module):
    def __init__(self, item_index):
        super(SelectItem, self).__init__()
        self._name = 'selectitem'
        self.item_index = item_index

    def forward(self, inputs):
        return inputs[self.item_index]
```

```python
#input size is (10,9)
class duo_lstm(torch.nn.Module):
    def __init__(self):
        super(duo_lstm, self).__init__()
        self.layer1 = torch.nn.LSTM(10, 15,3)
        self.h0 = torch.zeros(3, 1, 15)
        self.c0 = torch.zeros(3, 1, 15)
        self.select = SelectItem(0)

    def forward(self, x):
        y_pred, (h, _) = self.layer1(x, (self.h0, self.c0))
        print(h.shape)
        return self.select(y_pred)
```

```python


class test_seq(torch.nn.Module):
    def __init__(self):
        super(test_seq, self).__init__()
        self.net = torch.nn.Sequential(
            torch.nn.Linear(10, 64),
            torch.nn.ReLU(),
            torch.nn.LSTM(64, 256, 20),# 20 layers of lstm
            SelectItem(0),
            torch.nn.GRU(256, 128, 10),# 10 layers of gru
            SelectItem(0),
            torch.nn.Linear(128, 2),
            torch.nn.Softmax(dim=1)
        )

    def forward(self, x):
        y_pred = self.net(x)
        return y_pred

model = test_seq()
out = model(torch.randn(1, 10))
out
```
